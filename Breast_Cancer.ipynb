```python
y = df['diagnosis']
X = df.drop('diagnosis',axis=1)
```

Splitting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/9, random_state=252)
```

Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.fit_transform(X_test)
```

Creating Classification Models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

Rf = RandomForestClassifier()
Rf.fit(X_train, y_train)
pred_train = Rf.predict(X_test)
score= accuracy_score(y_test,pred_train)
print(score)
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
param_grid = {'random_state' : np.arange(0,300,2)}
RFC = RandomForestClassifier()
RFC_cv = GridSearchCV(RFC,param_grid,cv=10)
RFC_cv.fit(X_train, y_train)
RFC_cv.best_params_
RFC_cv.best_score_
```

```python
print('best random state : ',RFC_cv.best_params_)
print('best score : ',RFC_cv.best_score_)
```
